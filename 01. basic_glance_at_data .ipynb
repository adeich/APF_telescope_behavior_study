```python
import os
from os import path
import pandas
from datetime import datetime, timedelta
import numpy as np
from astropy.time import Time
import pytz
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import calendar


%matplotlib inline
```

```python
data = pandas.read_csv('final_data.csv')

data['delta_TAVERAGE'] = data['TAVERAGE'] - data['night_start_temperature']
data['delta_TM1S210'] = data['TM1S210'] - data['night_start_TM1S210']
data['delta_TM2CAIR'] = data['TM2CAIR'] - data['night_start_TM2CAIR']
print(data.columns)
print(data['TAVERAGE'][0], data['night_start_temperature'][0], data['delta_TAVERAGE'][0])

# clean data of bad temperature records.
data = data.mask(data['TAVERAGE'] == 381.03)
data = data.mask(data['stellar_activity1'] == -1)
data = data.mask(data['AZENCVEL'] < -5000)
data = data.mask(data['stellar_activity1'] < 0)
data = data.mask(data['ELFLWERR'] > 50000)
data = data.mask(data['stellar_activity1'] > 0.26)


```

```python
plot_df = pandas.DataFrame([['HD10700_APF.vels', 'b'], ['HD185144_APF.vels', 'g'], 
                            ['HD9407_APF.vels', 'r']], columns=['starID', 'color'])

def make_time_dist(ax, bin_edges, starID=None, starcolor=None):
    width = 2000000 * 0.7
    for i, row in plot_df.iterrows():
        starID = row['starID']
        time = data.loc[data['starID'] == starID].EpocSeconds
        velocity = data.loc[data['starID'] == starID].velocity
    
        cut, bins = pandas.cut(time, bin_edges, retbins=True) 
        analysis = velocity.groupby(cut).agg(['mean', 'std'])
        


        mean_plus_std = analysis['mean'] + analysis['std']
        mean_minus_std = analysis['mean'] - analysis['std']
    
        x_vals = bins[:-1] + 0.5 * (-bins[:-1] + bins[1:]) + i * width*2
        ax.bar(x_vals, height = 2 * analysis['std'], width=width,
                    bottom = mean_minus_std, color = row['color'], alpha=0.6, align='center',
                      label='_nolegend_')
        
        ax.plot(x_vals, analysis['mean'],
                c = row['color'], label=starID)
       
    ax.grid()
    ax.legend(loc='upper left')
    
    


figure(num=None, figsize=(11, 11), dpi=80, facecolor='w', edgecolor='k')
opacity = 0.5
shape = 'circle'
fig, axs = plt.subplots(4, 1)
axs[0].set_ylabel("Velocity")
axs[0].grid()
axs[0].scatter(data.EpocSeconds, data.velocity, s=2)
axs[0].set_title('Velocity measurements over time')
axs[1].hist(data.EpocSeconds, label='all stars'); axs[1].set_ylabel('n exposures')
axs[1].set_title('Measurement counts over time')
#axs[2].scatter(data.EpocSeconds, [])
axs[2].scatter(data.loc[data['starID'] == plot_df['starID'][0]].EpocSeconds, 
               data.loc[data['starID'] == plot_df['starID'][0]].velocity,
                s=15, c=plot_df['color'][0], label='HD10700', alpha=opacity)
axs[2].scatter(data.loc[data['starID'] == plot_df['starID'][1]].EpocSeconds, 
               data.loc[data['starID'] == plot_df['starID'][1]].velocity, 
                s=15, c=plot_df['color'][1], label='HD185144', alpha=opacity)
axs[2].scatter(data.loc[data['starID'] == plot_df['starID'][2]].EpocSeconds, 
               data.loc[data['starID'] == plot_df['starID'][2]].velocity, 
                s=15, c=plot_df['color'][2], label='HD9407', alpha=opacity)
axs[2].legend(loc='upper left');
axs[2].set_ylabel('velocity')
axs[2].grid()
axs[2].set_title('Velocity over time, distinguished by star')
axs[3].set_xlabel('epoc seconds')
make_time_dist(axs[3], bin_edges=[1.4e9, 1.428e9, 1.45e9, 1.485e9, 1.52e9, 1.55e9])
axs[3].set_ylabel('velocity')
axs[3].set_ylim(-10,10)
axs[3].set_title('Each star\'s avg velocity (center of bar)\n and variance of velocity (width of bar)')

fig.set_figheight(10)
fig.set_figwidth(10)
fig.tight_layout()
```

```python
def plot_temps_over_time():
    figure(num=None, figsize=(11, 11), dpi=80, facecolor='w', edgecolor='k')
    opacity = 0.5
    shape = 'circle'
    fig, axs = plt.subplots(4, 1)
    
    axs[0].set_ylabel("Temperature (C)")
    axs[0].grid()
    axs[0].scatter(data.EpocSeconds, data['TAVERAGE'], 
                   s=2, label='TAVERAGE')
    axs[0].set_title('Temperature at exposure time')
    
    axs[1].set_ylabel("Temperature (C)")
    axs[1].grid()
    axs[1].scatter(data.EpocSeconds, data['delta_TAVERAGE'], 
                   s=2, label='delta_TAVERAGE')
    axs[1].set_title('Change in temperature (TAVERAGE) between 17:00 and exposure time')
    
    axs[2].set_ylabel("Temperature (C)")
    axs[2].grid()
    axs[2].scatter(data.EpocSeconds, data['delta_TM1S210'], 
                   s=2, label='delta_TM1S210')
    axs[2].set_title('Change in temperature (TM1S210) between 17:00 and exposure time')
    
    axs[3].set_ylabel("Temperature (C)")
    axs[3].grid()
    axs[3].scatter(data.EpocSeconds, data['delta_TM2CAIR'], 
                   s=2, label='delta_TM2CAIR')
    axs[3].set_title('Change in temperature (TM2CAIR) between 17:00 and exposure time')
    
    fig.set_figheight(10)
    fig.set_figwidth(10)
    fig.tight_layout()
    
plot_temps_over_time()
```

```python
# clean data of bad temperature records.
data = data.mask(data['TAVERAGE'] == 381.03)
data = data.mask(data['stellar_activity1'] == -1)
data = data.mask(data['AZENCVEL'] < -5000)
data = data.mask(data['stellar_activity1'] < 0)
data = data.mask(data['ELFLWERR'] > 50000)
data = data.mask(data['stellar_activity1'] > 0.26)

print(len(data))

def make_correlations_exploring_plot(starID=None, limit=18):
    rows = int(np.floor(limit/3))
    fig, axarr = plt.subplots(rows, 3)

    if starID:
        masked_data = data.mask(data['starID'] == starID)
    else:
        masked_data = data
        starID = 'All_stars'
    
    
    correlations = masked_data.corr()['velocity']
    sorted_correlations_list = np.abs(correlations).sort_values(0, ascending=False)[3:-2]
    index = sorted_correlations_list.index
    
    sorted_correlations_list = sorted_correlations_list[:rows*3]

    for i, (name, correlation) in enumerate(sorted_correlations_list.iteritems()):
        row = int(np.floor(i/3)); col = i%3
        ax = axarr[row, col]
        ax.scatter(masked_data[name], masked_data['velocity'], s=1)
        if col == 0:
            ax.set_ylabel('velocity')
        #np.coerrcoeff
        #p = np.polyfit(data[name], data['velocity'], deg=1)
        #print(name, p)
        #x = np.arange(data[name].min(), data[name].max(), 100)
        #ax.plot(x, np.polyval(p, x), c='r')
        ax.set_xlabel(name)
        ax.set_title('{}: {}'.format(name, round(correlations[name], 2)))
        ax.autoscale()
    
    fig.set_figheight(15)
    fig.set_figwidth(10)
    fig.suptitle('Highest-to-lowest correlations with velocity\n {}'.format(starID),
                 size=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.savefig('{}_velocity_correlations.png'.format(starID))

for starID in plot_df['starID']:
    make_correlations_exploring_plot(starID = starID)
make_correlations_exploring_plot()
```
